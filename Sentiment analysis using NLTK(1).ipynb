#  Stage 2 : Tokenization

```python
clean_text =[]
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt')
```

```python
#Sentence Tokenize
sent_tok=[]
for sent in lower_text:
    sent=sent_tokenize(sent)
    sent_tok.append(sent)
```

```python
sent_tok
```

```python
#Word Tokenize
clean_text=[word_tokenize(i) for i in lower_text]
```

```python
clean_text
```

```python
import re    #Importing Regular Expression
clean_text_2=[]
for words in clean_text:
    clean=[]
    for w in words:
        res=re.sub(r'[^\w\s]',"",w) #Removing Special Characters
        if res!="":
            clean.append(res)
        clean_text_2.append(clean)
```

```python
clean_text_2
```

# Stage 3 : Removing Stop words

Stop words are words which are most commenly repeated.

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
clean_text_3=[]
for words in clean_text_2:
    w=[]
    for word in words:
        if not word in stopwords.words('english'):
            w.append(word)
        clean_text_3.append(w)
```

```python
clean_text_3
```

```python
X=df[: ,1: ]
X
```

```python
y=df[: ,0]
y
```

```python
X_train,y_train,X_test,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
def prior_probab(y_train,label):
    m=y_train.shape[0]
    s=np.sum(y_train == label)
    return m/s
```

```python
def cond_probab(X_train,y_train,feature_col,feature_val,label):
    X_filtered = X_train[y_train == label]
    num=np.sum(X_filtered[: ,feature_col]==feature_val)
    denom = X_filtered.shape[0]
    return float(num/denom)
```

```python
def predict(X_train,y_train,X_test):
    classes=np.unique(y_train)
    n_features = X_train.shape[1]
    posterior_probab =[]
    for label in classes:
        likelyhood=1.0
        for fea in range(n_features):
            cond = cond_probab(X_train,y_train,fea,X_test[fea],label)
            likelyhood=likelyhood*cond
        prior = prior_probab(y_train,label)
        post=likelyhood*prior
        posterior_probab.append(post)
        pred=np.argmax(posterior_probab)
        return pred
    
```

```python
def accuracy(X_train,y_train,X_test,y_test):
    pred=[]
    for i in range(X_test.shape[0]):
        p = predict(X_train,y_train,X_test[i])
        pred.append(p)
    y_pred=np.array(pred)
    acc=np.sum(y_pred==y_test)/y_pred.shape[0]
    return acc
```
